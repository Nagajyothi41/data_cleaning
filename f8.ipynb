```python
import matplotlib.pyplot as plt
import numpy as np
mark1=np.array([2,4,5,7,9,10,11,13,14,15])
mark2=np.array([6,7,8,10,11,12,13,14,15,16])
plt.hist([mark1,mark2],bins=10,stacked=True,color=["skyblue","salmon"],edgecolor="blue")
plt.xlabel("marks")
plt.ylabel("count")
plt.title("stacked histogram with multiple dataset")
plt.legend(["mark dataset1","mark dataset 2"])
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
exam_score=np.random.randint(0,100,150)
plt.hist(exam_score,bins=20,cumulative=True,color="skyblue",edgecolor="black")
plt.xlabel("exam scores")
plt.ylabel("cumulative number of students")
plt.title("cumulative histogram of exam score")
plt.show()
```
